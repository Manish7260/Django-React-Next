```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data_set = pd.read_csv('Position_Salaries.csv')
```

```python
data_set
```

```python
x = data_set.iloc[:,1:2].values
y = data_set.iloc[:,-1].values
```

```python
x
```

```python
y
```

```python
from sklearn.linear_model import LinearRegression
```

```python
li_reg = LinearRegression()
```

```python
li_reg.fit(x,y)
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly_reg = PolynomialFeatures(degree=2)
```

```python
x_poly = poly_reg.fit_transform(x)
```

```python
li_reg_2 = LinearRegression()
```

```python
li_reg_2.fit(x_poly, y)
```

```python
plt.scatter(x,y,color="blue")
plt.plot(x, li_reg.predict(x), color="red")
plt.title("Bluff Detection Model (Linear Regression Model)")
plt.xlabel("Position Level")
plt.ylabel("Salary")
plt.plot()
```

```python
plt.scatter(x,y,color="blue")
plt.plot(x, li_reg_2.predict(poly_reg.fit_transform(x)), color="red")
plt.title("Bluff Detection Model (Linear Regression Model)")
plt.xlabel("Position Level")
plt.ylabel("Salary")
plt.plot()
```

```python
poly_reg = PolynomialFeatures(degree=3)
```

```python
x_poly= poly_reg.fit_transform(x)
```

```python
li_reg_3 = LinearRegression()
```

```python
li_reg_3.fit(x_poly,y)
```

```python
plt.scatter(x,y,color="blue")
plt.plot(x, li_reg_3.predict(poly_reg.fit_transform(x)), color="red")
plt.title("Bluff Detection Model (Linear Regression Model)")
plt.xlabel("Position Level")
plt.ylabel("Salary")
plt.plot()
```
