```python
import pandas as pd

dataset = pd.read_csv('Fake_Real_Data.csv')

print(dataset.shape)

dataset.head()
```

```python
dataset['label'].value_counts()
```

```python
dataset['label_num'] = dataset['label'].map({"Fake" : 0, "Real":1})
```

```python
dataset.head()
```

```python
!python -m spacy download en_core_web_lg
```

```python
import spacy

nlp = spacy.load('en_core_web_lg')
```

```python
dataset['vector'] = dataset['Text'].apply(lambda text: nlp(text).vector)  
```

```python
dataset.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    dataset.vector.values, 
    dataset.label_num, 
    test_size=0.2, 
    random_state=2022
)
```

```python
X_train
```

```python
import numpy as np

X_train_2d = np.stack(X_train)
X_test_2d = np.stack(X_test)
```

```python
X_train_2d
```

```python
X_test_2d
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

scaled_train = scaler.fit_transform(X_train_2d)
scaled_test = scaler.fit_transform(X_test_2d)

classifier = MultinomialNB()

classifier.fit(scaled_train, y_train)
```

```python
y_pred = classifier.predict(scaled_test)
```

```python
y_pred
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred))
```

```python
from sklearn.neighbors import KNeighborsClassifier

classifier = KNeighborsClassifier(n_neighbors=5, metric='euclidean')

classifier.fit(scaled_train, y_train)
```

```python
y_pred = classifier.predict(scaled_test)

print(classification_report(y_test, y_pred))
```

```python
#finally print the confusion matrix for the best model
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
cm

from matplotlib import pyplot as plt
import seaborn as sn
plt.figure(figsize = (10,7))
sn.heatmap(cm, annot=True, fmt='d')
plt.xlabel('Prediction')
plt.ylabel('Truth')
```
